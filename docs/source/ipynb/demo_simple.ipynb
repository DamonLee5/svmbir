# A simple demo uses svmbir for reconstruction.

```python
import numpy as np
import matplotlib.pyplot as plt
import svmbir
```

## Load Sinogram Data

```python
# Load data
sino = np.load('../../../demo/sinodata.npy')
# Generate angles list
(num_views, num_slices, num_channels) = sino.shape
angles = np.linspace(0, np.pi, num_views, endpoint=False)
```

## Simple Reconstruction

```python
# center_offset accounts for the center of rotation being offset from the center of the detector by 6 detector channels
x = svmbir.recon(sino, angles, center_offset=-6)
```

```python
## display output #################################
imgplot = plt.imshow(x[0])
imgplot.set_cmap('gray')
plt.colorbar()
plt.show()
plt.close()
```

## Sharper Reconstruction

```python
x_sharp = svmbir.recon(sino, angles, sharpness=2.0, center_offset=-6)
```

```python
## display output #################################
imgplot = plt.imshow(x_sharp[0])
imgplot.set_cmap('gray')
plt.colorbar()
plt.show()
plt.close()
```
